**Step 1: Building your data set.**

Use NumPy to:

1. Randomly generate 100 numbers between 0 and 1 and save them to an
array ’x’. These are your input values.
2. Create an array ’t’. For each entry x[i] in x, calculate x[i]**3-x[i]**2
and save the results to t[i]. These are your targets.


```python
import numpy as np
import matplotlib.pyplot as plt

#generate 100 random numbers between 0 and 1
#save them to an array "x"
x = np.random.rand(100)
#sort the 100 random numbers in x - for easier visualization with a graph
x = np.sort(x)
#to create array t, calculate x[i]**3-x[i]**2 and save result in t[i]
t = x**3 - x**2


#plotting the results
plt.scatter(x,t)
plt.plot(x,x**3-x**2)
plt.show
```
